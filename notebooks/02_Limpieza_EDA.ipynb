```python
correlaciones=df.corr()                          
print(correlaciones["CLASS_LABEL"].sort_values()[:5], "\n", correlaciones["CLASS_LABEL"].sort_values()[-5:])        
```

```python
plt.scatter(df['FrequentDomainNameMismatch'], df['CLASS_LABEL'])
plt.xlabel('FrequentDomainNameMismatch')
plt.ylabel('CLASS_LABEL')                                                 # ? Mirar las grÃ¡ficas de la guÃ­a.
plt.title('Diagrama de DispersiÃ³n');
```

```python
# Pruebo a filtrar para mostrar con un mapa de calor las variables que estÃ¡n mÃ¡s correlacionadas
```

```python
filtro = correlaciones[(correlaciones['CLASS_LABEL']>0.15) | (correlaciones['CLASS_LABEL']<-0.15)]
mantener = filtro["CLASS_LABEL"].index
columnas = [col for col in df.columns if col not in mantener]
```

```python
df = df.drop(columns=columnas)
```

```python
df.info()
```

```python
plt.figure(figsize=(13,8))
sns.heatmap(df.corr(), annot=True, linewidths=2, linecolor="black");
```

```python
fig, axes = plt.subplots(8, 2, figsize=(15, 20))  

No_Phishing = df[df["CLASS_LABEL"] == 0]
Phishing = df[df["CLASS_LABEL"] == 1]

# Itero sobre los ejes y las columnas del DF
for i, ax in enumerate(axes.flatten()):
    if i < len(df.columns):
        col_name = df.columns[i]
        ax.hist(No_Phishing[col_name], bins=40, color='r', alpha=0.5, label='No_Phishing')
        ax.hist(Phishing[col_name], bins=40, color='g', alpha=0.5, label='Phishing')
        ax.set_title(col_name)
        ax.legend()

plt.tight_layout();
```
